```python
def add_remaining_useful_life(df):
    grouped_by_unit = df.groupby(by="unit_no")
    max_cycle = grouped_by_unit["time_cycles"].max()
    
    result_frame = df.merge(max_cycle.to_frame(name='max_cycle'), left_on='unit_no', right_index=True)
    
    remaining_useful_life = result_frame["max_cycle"] - result_frame["time_cycles"]
    result_frame["RUL"] = remaining_useful_life
    
    result_frame = result_frame.drop("max_cycle", axis=1)
    return result_frame
```

```python
train = add_remaining_useful_life(train)
train.head()
```

```python
#-- Amount of variation 

#Drop variables with zero variation
var_data = train[sensor_names].var()
print(var_data)
if False: # If True use the standard deviation. Use when not scaled
  var_data = np.sqrt(var_data) 

plt.plot(np.arange(len(var_data)), var_data.sort_values(ascending = False))
plt.xlabel('Number of Features')
plt.ylabel('Variance')
```

## Plot the distribution of RUL and sensor values

```python
#Distribution of target variable, RUL
df_max_rul = train[['unit_no', 'RUL']].groupby('unit_no').max().reset_index()
df_max_rul['RUL'].hist(bins=15, figsize = (15,7))
```

```python
def plot_sensor_distribution(sensor_name):
    plt.figure(figsize=(13,5))
    for i in train['unit_no'].unique():
        #if (i % 10 == 0):
        if(i == 1):
            train[sensor_name][train['unit_no']==i].hist(bins=15, figsize = (10,5))
    plt.ylabel(sensor_name)
    plt.show()
```

```python
#Plot distribution of sensors for 1 engine
for sensor_name in sensor_names:
    plot_sensor_distribution(sensor_name)
```

## Plot sensor values over time/cycles

```python
def plot_sensor(sensor_name):
    plt.figure(figsize=(10,5))
    for i in train['unit_no'].unique():
        if (i % 10 == 0):
        #if(i == 1):
            plt.plot('RUL', sensor_name, 
                     data=train[train['unit_no']==i])
    plt.xlim(250, 0)  # reverse the x-axis so RUL counts down to zero
    plt.xticks(np.arange(0, 275, 25))
    plt.ylabel(sensor_name)
    plt.xlabel('Remaining Use fulLife')
    plt.show()
```

```python
for sensor_name in sensor_names:
    plot_sensor(sensor_name)
```

## Stationarity

```python
# Check if the data is stationary
# Use Augmented Dickey Fuller test, find p-value, if p-value < 0.05 -> data is stationary
# Try using differencing to improve p-value???

from statsmodels.tsa.stattools import adfuller

def check_stationarity(df, sensor):
    test_sample = df[sensor][df['unit_no'] == 1]
    ad_test = adfuller(test_sample, maxlag=1)
    print('Sensor: {}\nADFuller test statistics: {}\nP-Value: {}\n\n'.format(sensor,ad_test[0], ad_test[1]))
```

```python
for s in s_sensor_names:
    check_stationarity(selected_train, s)
```
