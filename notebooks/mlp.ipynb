```python
import azureml.core
from azureml.core import Workspace, Dataset

#Load workspace from the saved config file
try:
     ws = Workspace.from_config()
     print("Ready to use Azure ML SDK {} to work with {}".format(azureml.core.VERSION, ws.name))
except:
     print("Could not load AML workspace")
```

```python
import numpy as np
import pandas as pd
import math

import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from keras.callbacks import EarlyStopping

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
# Training data
dataset = Dataset.get_by_id(ws, id='2035914e-063b-434f-89b0-b0e391b22270')
data = dataset.to_pandas_dataframe()

x_data = data.copy()
x_data['target_45d'] = np.where(x_data['RUL'] <= 45, 1, 0)
x_data = x_data.drop(['sensor_22','sensor_23','sensor_24','sensor_25','sensor_26','sensor_27', 'RUL'], axis = 1)
y_data = pd.DataFrame(x_data.pop('target_45d').values.reshape(-1,1))
```

```python
# Testing data
RUL_df = pd.read_csv('..\data/RUL_test.csv')
dataset = Dataset.get_by_id(ws, id='74fd6c48-508d-46b2-9428-09a7057c7ede')
temp_data = dataset.to_pandas_dataframe()
temp_data = temp_data.drop(['sensor_22','sensor_23','sensor_24','sensor_25','sensor_26','sensor_27'], axis = 1)

x_test_data = pd.DataFrame()

for unique_value in temp_data['engine_no'].unique():
    df1 = temp_data[temp_data['engine_no']==unique_value]
    idx = int(45 - RUL_df['RUL'][RUL_df['engine_no']==unique_value])
    if idx >= 0:
        df1['target_45d'] = [0] * (len(df1)-idx) + [1] * idx
    else:
        df1['target_45d'] = [0] * len(df1)
    x_test_data = pd.concat([x_test_data,df1])

y_test_data = pd.DataFrame(x_test_data.pop('target_45d').values.reshape(-1,1))
```

```python
#data.head(500)
x_signals = x_data.shape[1]
y_signals = y_data.shape[1]
num_train = x_data.shape[0]
print('Number of features: {}\nNumber of targets: {}\nNumber of samples: {}'.format(x_signals, y_signals, num_train))
```

```python
sc = MinMaxScaler(feature_range = (0,1))
x_data[x_data.columns[2:26]] = sc.fit_transform(x_data[x_data.columns[2:26]])
x_test_data[x_test_data.columns[2:26]] = sc.fit_transform(x_test_data[x_test_data.columns[2:26]])
```

```python
enc = OneHotEncoder()
y_data = enc.fit_transform(y_data).toarray()
y_test_data = enc.fit_transform(y_test_data).toarray()
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_data, y_data, test_size=0.2, random_state=2)

print(x_train.shape)
print(x_val.shape)
print(x_test_data.shape)
```

```python
model = Sequential()
model.add(Dense(128, input_dim=26, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(2, activation='softmax'))

model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
model.summary()
```

```python
early_stopping = EarlyStopping(monitor = 'val_loss', patience = 10, verbose = 2, restore_best_weights = True)

model.fit(x_train, y_train, epochs = 1000, batch_size = 100, validation_data = (x_val, y_val), 
          verbose = 2,
          callbacks = [early_stopping])
```

```python
score = model.evaluate(x_test_data, y_test_data, verbose=2)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
